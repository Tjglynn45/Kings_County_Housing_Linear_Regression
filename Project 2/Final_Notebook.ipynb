# King County Housing Data

For this project, we are analyzing King County housing data from Washington State. We are trying to create a predictive model for housing costs. 

# Obtaining and Inputting the Data

```python
#importing libraries for all the data cleaning analysis and regression
import numpy as np
import pandas as pd
from scipy import stats
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import statsmodels.api as sm
from sklearn import preprocessing
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
#import the data into a dataframe
df = pd.read_csv('kc_house_data_train.csv',index_col='Unnamed: 0')
column_names = df.columns
```

# Exploratory Data Analysis

We are going to take a look at the data to start doing some data analysis and create data visualizations.

```python
#Summary of the data
df.info()
```

```python
#Summary of the statistics of the data
df.describe()
```

```python
#Dropping the Id as it isnt really relevant
df = df.drop(['id'], axis=1)
```

```python
#Create a heatmap to see the correlation between price and DF
plt.figure(figsize=(10,10))
corr=df.corr()
sns.heatmap(corr,center=0)
```

```python
#Trying to see if any values are highly correlated
abs(corr) >0.75
```

```python
#We drop the highly correlated columns
#sqft is more valueable than bathrooms and is directly correlates with sqft above
#grade is similar to to condition so we drop grade
df.drop(['bathrooms'], axis=1, inplace=True)
df.drop(['grade'], axis=1, inplace=True)
df.drop(['sqft_above'], axis=1, inplace=True)
```

```python
#Checking for outliers
df.describe()
```

```python
# Looking at the summary stats of bedrooms and also a boxplot
# we can see there is one entry that has 33 bedrooms
plt.figure(figsize=(10,5))
sns.boxplot(df.bedrooms)
```

```python
# That outlier may be mistake based on sqft of the house so we will change it to the more common room number of 3
df['bedrooms'] = np.where(df['bedrooms'] == 33, 3, df['bedrooms'] )
```

```python
#Change the date column using the datetime format
df.date = pd.to_datetime(df.date)
df['month'] = df.date.dt.month
df.month =df.month.replace(12,0)
bins = [0,3,6,9,11]
bins_date = pd.cut(df['month'], bins, include_lowest=True, labels=['Winter','Summer','Spring','Fall'])
bins_date = bins_date.cat.as_unordered()
```

```python
#viewing the houses sold based on season
bins_date.value_counts().plot(kind='bar')
```

```python
#Group houses built per 25 years from 1900-2015
bins = [1900, 1925, 1950, 1975, 2000, 2015]
bins_yr_built = pd.cut(df['yr_built'], bins, include_lowest=True, labels = ('1900-1925', '1926-1950', '1951-1975', '1976-2000', '2001-2015'))
bins_yr_built = bins_yr_built.cat.as_unordered()
```

```python
#Plot the amount of houses build per 25 year groups
df_plot_me = pd.DataFrame({'Year_bin': bins_yr_built.values, 'Count':1})
df_plot_me.groupby('Year_bin').sum().plot(kind='bar')
```

```python
#Make renovated 1 if its been renovated and 0 if it hasnt
df['renovated']=df['yr_renovated'].apply(lambda x:1 if x>0 else 0)
```

```python
#if the house has a basement 1,if it hasnt 0
df['has_basement'] = df['sqft_basement'].apply(lambda x:1 if x>0 else 0)
```

```python
#Grouping the dataframe by categorical and numerical to have 2 different dataframes
categorical = df[['month','view','condition','yr_built','yr_renovated','zipcode','renovated','has_basement']].copy()
numerical = df[['price','bedrooms','sqft_living','floors','sqft_lot','sqft_basement','lat','long','sqft_living15','sqft_lot15']].copy()
```

```python
#Making the column categorical as viewed or not viewed instead of number of views
bins = [0, 1, 4] 
bins_view = pd.cut(categorical['view'], bins, include_lowest=True, labels=['not_viewed', 'viewed'])
bins_view = bins_view.cat.as_unordered()
```

```python
categorical = categorical.drop(['view', 'month', 'yr_built','yr_renovated'], axis=1)
categorical = pd.concat([categorical, bins_date, bins_view, bins_yr_built,], axis=1)
```

```python
#Turning them into str so they can be dummied out
categorical['month'] = categorical['month'].astype('str')
categorical['view'] = categorical['view'].astype('str')
categorical['condition'] = categorical['condition'].astype('str')
categorical['zipcode'] = categorical['zipcode'].astype('str')
categorical['renovated'] = categorical['renovated'].astype('str')
categorical['has_basement'] = categorical['has_basement'].astype('str')
categorical['yr_built'] = categorical['yr_built'].astype('str')
```

```python
categorical = pd.get_dummies(categorical)
categorical.info()
```

```python
# Creating polynomial features for sqft_living and bedrooms
poly = PolynomialFeatures(degree=2,include_bias=False)
features = ['sqft_living','bedrooms']
df_features = df[features]
poly_data = poly.fit_transform(df_features)
poly_columns = poly.get_feature_names(df_features.columns)
df_poly=pd.DataFrame(poly_data,columns=poly_columns)
df_poly.head()
```

```python
#combine the numerical and categorical dataframes together to make a final DF
final = pd.concat([numerical, categorical],axis=1)
final.head()
```
